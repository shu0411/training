```python
#C
import random

A = ""
for i in range(100):
    #10**8A
    A += str(random.randint(1, 10 ** 8)) + " "
#
A = A[:-1]
print(A)

B = ""
for i in range(100):
    #10**8B
    B += str(random.randint(1, 10 ** 8)) + " "
#
B = B[:-1]
print(B)

#C
C = ""
for i in range(100):
    #10**8C
    C += str(random.randint(1, 10 ** 8)) + " "
#
C = C[:-1]
print(C)

#D
D = ""
for i in range(2 * (10 ** 5)):
    #10**8D
    D += str(random.randint(1, 3 * (10 ** 8))) + " "
#
D = D[:-1]
print(D)

```
