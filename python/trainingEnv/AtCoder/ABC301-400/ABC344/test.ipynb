```python
#C
import random

A = ""
for i in range(100):
    #10**8以下のランダムな数を半角スペース区切りでAに追加
    A += str(random.randint(1, 10 ** 8)) + " "
#最後のスペースを削除
A = A[:-1]
print(A)

B = ""
for i in range(100):
    #10**8以下のランダムな数を半角スペース区切りでBに追加
    B += str(random.randint(1, 10 ** 8)) + " "
#最後のスペースを削除
B = B[:-1]
print(B)

#C
C = ""
for i in range(100):
    #10**8以下のランダムな数を半角スペース区切りでCに追加
    C += str(random.randint(1, 10 ** 8)) + " "
#最後のスペースを削除
C = C[:-1]
print(C)

#D
D = ""
for i in range(2 * (10 ** 5)):
    #10**8以下のランダムな数を半角スペース区切りでDに追加
    D += str(random.randint(1, 3 * (10 ** 8))) + " "
#最後のスペースを削除
D = D[:-1]
print(D)

```
