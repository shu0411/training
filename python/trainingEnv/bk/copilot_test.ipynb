```python
from itertools import permutations

#標準入力から4つの数字を取得
a, b, c, d = map(int, input().split())

out = '解なし'

#4つの数字を加減乗除して10を作る
for op1, op2, op3 in permutations(['+', '-', '*', '/'], 3):
    try:
        if eval(f"{a}{op1}{b}{op2}{c}{op3}{d}") == 10:
            out = f"{a}{op1}{b}{op2}{c}{op3}{d}"
            break
    except ZeroDivisionError:
        pass

#1つ目の式の答えと残った2つの数字のうち2つを選んで加減乗除のいずれかで計算する
if out != '解なし':
    for op1, op2 in permutations(['+', '-', '*', '/'], 2):
        try:
            if eval(f"{out}{op1}{a}{op2}{b}") == 10:
                out = f"({out}){op1}{a}{op2}{b}"
                break
            elif eval(f"{out}{op1}{a}{op2}{c}") == 10:
                out = f"({out}){op1}{a}{op2}{c}"
                break
            elif eval(f"{out}{op1}{a}{op2}{d}") == 10:
                out = f"({out}){op1}{a}{op2}{d}"
                break
            elif eval(f"{out}{op1}{b}{op2}{c}") == 10:
                out = f"({out}){op1}{b}{op2}{c}"
                break
            elif eval(f"{out}{op1}{b}{op2}{d}") == 10:
                out = f"({out}){op1}{b}{op2}{d}"
                break
            elif eval(f"{out}{op1}{c}{op2}{d}") == 10:
                out = f"({out}){op1}{c}{op2}{d}"
                break
        except ZeroDivisionError:
            pass

#2つ目の式の答えと残った1つの数字を選んで加減乗除のいずれかで計算する
if out != '解なし':
    for op in ['+', '-', '*', '/']:
        try:
            if eval(f"{out}{op}{a}") == 10:
                out = f"({out}){op}{a}"
                break
            elif eval(f"{out}{op}{b}") == 10:
                out = f"({out}){op}{b}"
                break
            elif eval(f"{out}{op}{c}") == 10:
                out = f"({out}){op}{c}"
                break
            elif eval(f"{out}{op}{d}") == 10:
                out = f"({out}){op}{d}"
                break
        except ZeroDivisionError:
            pass

print(out)
```
